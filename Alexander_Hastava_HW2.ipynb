#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

You are at the age of 21 and curious about if drinking alcohol would cause liver disease. You contacted a clinic site and got the following information.

1. Past data tells you that 10% of patients entering the clinic have liver disease (**Prior**). 

2. Five percent of the clinicâ€™s patients are alcoholics (**Marginal Probability**). 

3. Among the patients diagnosed with liver disease, 15% are alcoholics (**Likelihood**). 

Now you wonder what is the probability of a patient from the clinic have liver disease given that the patient is alcoholic (**Posterior**)

### 1.1 

#### If A represents the event of being an alcoholic patient in the clinic and L represents the event of being a patient with liver disease in the clinic, write the bayesian representation (P(A|B)) in the following cells


Posteria = 

```julia
using LaTeXStrings
L"P(L|A) = \frac{P(A|L) \cdot P(L)}{P(A)}"
```

Prior = 

```julia
L"P(L) = 0.10"
```

Likelihood = 

```julia
L"P(A|L) = \frac{P(L|A) \cdot P(A)}{P(L)}"
```

Marginal Probability = 

```julia
L"P(A) = \sum_{L} P(A|L)P(L) + P(A|\neg L)P(\neg L)"
```

### 1.2 

#### Define all the information available and calculate the posterior probability

```julia
P_L = 0.10  # Prior probability of having liver disease
P_A = 0.05  # Marginal probability of being an alcoholic
P_A_given_L = 0.15  # Likelihood of being an alcoholic given liver disease

# Calculate the probability of not having liver disease
P_not_L = 1 - P_L

# Calculate the probability of being an alcoholic given no liver disease
P_A_given_not_L = (P_A - P_A_given_L * P_L) / P_not_L

# Calculate the posterior probability using Bayes' theorem
P_L_given_A = (P_A_given_L * P_L) / P_A

println("The posterior probability P(L|A) is: $(round(P_L_given_A, digits=4))")
```

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
P_C = 0.01  # Prior probability of having Covid
P_T_given_C = 0.653  # Sensitivity: Probability of testing positive given having Covid
P_T_given_not_C = 1 - 0.999  # 1 - Specificity: Probability of testing positive given not having Covid

# Calculate the probability of not having Covid
P_not_C = 1 - P_C

# Calculate the marginal probability of testing positive
P_T = P_T_given_C * P_C + P_T_given_not_C * P_not_C

println("The marginal probability P(T) is: $(round(P_T, digits=4))")
```
